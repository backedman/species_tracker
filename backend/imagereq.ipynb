```python
import requests
from io import BytesIO
from PIL import Image
```

```python
def get_img(taxa): 
    url = 'https://api.inaturalist.org/v1/taxa?taxon_id='+taxa+'&order=desc&order_by=observations_count'
    response = requests.get(url)
    ob = response.json()

    if ob.get("results", 'error') == 'error':
        return None
    
    if ob.get("total_results", 'error') == 0:
        return None

    # Print the response
    image = ob['results'][0]['default_photo']['url']
    response = requests.get(image)
    img = Image.open(BytesIO(response.content))
    return img
```

```python

img = get_img('4328')
display(img)

    
```
