```python
import pandas as pd
import numpy as np
```

```python
d = pd.read_csv("LPD_2024_public.csv")
d.iloc[0]
```

```python
d = d.drop(['Replicate', 'Included in LPR2024', 'Citation', 'Common_name', 'Order', 'Family', 'Genus', 'Species', 'Subspecies', 'IPBES_region', 'Region', 'IPBES_subregion', 'Latitude', 'Longitude', 'Specific_location', 'System', 'FW_biome', 'T_realm', 'T_biome', 'FW_realm', 'M_ocean', 'M_realm', 'M_biome'], axis=1)
```

```python
df = d[d.Country == 'United States']
df
```

```python
#deal with Estimated Wild Population 1000s
mask = df['Units'] == "Estimated wild population (1000's)"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x * 1000 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

#deal with SSB thousands of lbs
mask = df['Units'] == "SSB (Thousands of lbs)"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x * 1000 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "SSB (Thousands)"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x * 1000 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

df.loc[df['Units'] == "Number of individuals", 'Units'] = 'Population_value'
df.loc[df['Units'] == "Population estimate", 'Units'] = 'Population_value'

df
```

```python
df = df[df.Class != 'Actinopteri']
df = df[df.Class != 'Actinopterygii']
df = df[df.Class != 'Elasmobranchii']
df = df[df.Class != 'Amphibia']
df = df[df.Class != 'Reptilia']

mask = df['Units'] == "Breeding population estimate 1,000's of pairs"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x * 2 *1000 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "Individuals (thousands)"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x *1000 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "Number of Pairs"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x * 2 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "Number of nests per season (thousands)"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x * 2 *1000 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "Number of males"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x *2 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "Number of breeding females"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x *2 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "Number of females"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x *2 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'
```

```python

mask = df['Units'] == "Number of pairs"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x *2 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "Nesting pairs"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x *2 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "Number of nests per season"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x *2 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "Number of breeding birds"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x *2 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "Number of female nesters"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x *2 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

print(df['Units'].unique())
```

```python

mask = df['Units'] == "number of individuals/1000 trap nights"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x *1000 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "Number of nests"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x *2 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "Nest index"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x *2 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "number of nests/km2"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x *2 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "number of nests"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x *2 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "No. nests"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x *2 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "Number of breeding pairs"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x *2 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "Occupied Nests"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x *2 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "Number of burrows"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x *2 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'

mask = df['Units'] == "Breeding pairs"
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]
df.loc[mask, year_columns] = df.loc[mask, year_columns].map(lambda x: x *2 if pd.notna(x) else x)
df.loc[mask, 'Units'] = 'Population_value'


df.loc[df['Units'] != 'Abundance index', 'Units'] = 'Population_value'

unit_type_counts = df['Units'].value_counts()
filtered_counts = unit_type_counts[unit_type_counts > 3]
print(filtered_counts)
```

```python
pop_df = df[df['Units'] == 'Population_value']
ab_df = df[df['Units'] == 'Abundance index']
year_columns = [col for col in df.columns if col.isdigit() and 1950 <= int(col) <= 2020]

# Replace NaNs with 0s (for summing purposes)
pop_df[year_columns] = pop_df[year_columns].fillna(0)

# Group by binomial name (species) and sum population data for each year
combined_df = pop_df.groupby('Binomial')[year_columns].sum().reset_index()

# Replace 0s back to NaNs
combined_df[year_columns] = combined_df[year_columns].replace(0, float('nan'))
```
